```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


# A. Una función que calcule el precio total de la compra
type(articulos) #diccionario
type(compra)    #lista
def total_compra (articulos, compra):
    precio = 0
    for i in compra:
      precio += articulos[i]
    
    return precio  
 
precio = total_compra (articulos, compra)
print(f"El precio de la compra es : {precio} euros")


# B. Una función que calcule el precio total de la compra con un descuento del 10%
precio 
def descuento_10(precio):
  descuento = (precio * 10)/100
  descuento = precio - descuento

  return descuento

descuento = descuento_10(precio)
print(f'El precio con el descuento del 10% es de {descuento}')

# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€

def descuento_10(precio):
  if(precio > 100):
    descuento = (precio * 10)/100
    descuento = precio - descuento
  else:
    descuento = precio
  return descuento

descuento = descuento_10(precio)
print(f'El precio con el descuento del 10% es de {descuento}')

# D. Una función que calcule el precio total aplicando el IVA (21%)

def iva(precio):
  precioIva = (precio * 21)/100
  precioIva = precio + precioIva
  return precioIva

precioIva = iva(precio)
print(f'El precio con IVA es de {precioIva}')

# E.lista los artículos cuyo precio es superior a 20€

lista = list(filter(lambda k: k[1] > 20, articulos.items()))
print(f'Los artículos con precio superior a 20 son: {lista}')
```
